```python
import edge_tts
import asyncio
import pygame

cuento = """Había una vez una ballena muy, muy grande que se llamaba Berta. 
Berta era de un color azul brillante, como el océano donde vivía. 
Un día, mientras Berta cantaba una canción bajita bajo el sol, vio a alguien nuevo cerca de unas rocas."""

voz_elegida = "es-ES-AlvaroNeural" 
archivo_salida = "cuento.mp3"

async def crear_audiolibro():
    
    comunicador = edge_tts.Communicate(cuento, voz_elegida)
    await comunicador.save(archivo_salida)

await crear_audiolibro()

# Para .py
# asyncio.run(crear_audiolibro())

# Usamos pygame para reproducir el audio
def reproducir_audio(archivo):
    pygame.mixer.init()

    pygame.mixer.music.load(archivo)
    
    pygame.mixer.music.play()
    
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

reproducir_audio("cuento.mp3")
```
