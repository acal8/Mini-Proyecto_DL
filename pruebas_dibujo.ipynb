Sin ControlNet, el modelo Stable Diffusion generaría imágenes únicamente a partir del prompt, ignorando el dibujo que le pasamos del niño. ControlNet "congela" el modelo base y añade capas de control para aumentar la precisión, preservando formas, poses y colores mientras que se mejoran los detalles.

ControlNet de scribble para mejorar el dibujo del niño

Stable Diffusion is a latent text-to-image diffusion model that generates high-resolution images based on text prompts. It uses a frozen CLIP ViT-L/14 text encoder to condition the model on text prompts and has an 860M UNet and 123M text encoder.
